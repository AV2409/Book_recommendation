```python
RAT_NAME = ratings.merge(books,on='ISBN')
```

```python
nuMRating_df = RAT_NAME.groupby('Book-Title').count()['Book-Rating'].reset_index()
nuMRating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
nuMRating_df
```

```python
avGRating_df = RAT_NAME.groupby('Book-Title').agg({'Book-Rating': 'mean'}).reset_index()
avGRating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avGRating_df
```

```python
x = RAT_NAME.groupby('User-ID').count()['Book-Rating'] > 200
INTELLIGENT = x[x].index
```

```python
filtered = RAT_NAME[RAT_NAME['User-ID'].isin(INTELLIGENT)]
```

```python
y = filtered.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final = filtered[filtered['Book-Title'].isin(famous_books)]
pt = final.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarityy = cosine_similarity(pt)
similarityy.shape
```

```python
def recommend(book_name):
    if book_name not in pt.index:
        return "Book not found in the dataset."

    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarityy[index])),key=lambda x:x[1],reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('1984')
# pt.index[545]
```
